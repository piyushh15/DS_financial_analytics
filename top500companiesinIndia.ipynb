```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans

df['Mar Cap - Crore'] = pd.to_numeric(df['Mar Cap - Crore'], errors='coerce')
df['Sales Qtr - Crore'] = pd.to_numeric(df['Sales Qtr - Crore'], errors='coerce')


df.dropna(subset=['Mar Cap - Crore', 'Sales Qtr - Crore'], inplace=True)

# Correlation Analysis
correlation, _ = pearsonr(df['Mar Cap - Crore'], df['Sales Qtr - Crore'])
print(f'Correlation coefficient: {correlation}')

# Linear Regression
X = df[['Mar Cap - Crore']]
y = df['Sales Qtr - Crore']
model = LinearRegression()
model.fit(X, y)
df['Predicted Sales Qtr - Crore'] = model.predict(X)

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Mar Cap - Crore', y='Sales Qtr - Crore', data=df, label='Actual')
sns.lineplot(x='Mar Cap - Crore', y='Predicted Sales Qtr - Crore', data=df, color='red', label='Predicted')
plt.title('Mar Cap vs Sales Qtr with Linear Regression')
plt.xlabel('Mar Cap (Crore)')
plt.ylabel('Sales Qtr (Crore)')
plt.legend()
plt.show()

# Clustering Analysis
kmeans = KMeans(n_clusters=3)
df['Cluster'] = kmeans.fit_predict(df[['Mar Cap - Crore', 'Sales Qtr - Crore']])

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Mar Cap - Crore', y='Sales Qtr - Crore', hue='Cluster', data=df, palette='viridis')
plt.title('Clustering Analysis of Mar Cap and Sales Qtr')
plt.xlabel('Mar Cap (Crore)')
plt.ylabel('Sales Qtr (Crore)')
plt.show()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.pivot_table(index='Mar Cap - Crore', columns='Sales Qtr - Crore', aggfunc='size'), cmap='viridis')
plt.title('Heatmap of Mar Cap vs Sales Qtr')
plt.xlabel('Sales Qtr (Crore)')
plt.ylabel('Mar Cap (Crore)')
plt.show()
```
